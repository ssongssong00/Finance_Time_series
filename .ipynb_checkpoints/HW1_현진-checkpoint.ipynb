```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/금융시계열/QUARTERLY.csv")
```

```python
df.head()
```

```python
import math
```

```python
yt_indprod = []
```

```python
for i in range(212):
    if i == 0: 
        yt = None
    else: 
        yt = math.log(df['IndProd'][i]) - math.log(df['IndProd'][i-1])
        
    yt_indprod.append(yt)
df['yt_indprod'] = yt_indprod
```

```python
df
```

```python
import statsmodels.api as sm
```

```python
model = sm.tsa.arima.ARIMA(df['yt_indprod'],order = (1,0,0))
result = model.fit()
result.summary()
```
