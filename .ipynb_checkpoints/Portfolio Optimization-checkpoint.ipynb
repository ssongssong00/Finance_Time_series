```python
# 기본적인 패키지 로드
import numpy as np
import pandas as pd
from datetime import datetime

# 포트폴리오 최적화 패키지 로드
# pip install
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns

# yahoo finance 데이터 불러오기 패키지 로드
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()

# 시각화 패키지 로드
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# 종목 선정
Assets = ["MSFT","Gold", "AAPL", "TSLA", "META"]

# 수정 종가 정보를 불러올 빈 데이터프레임 생성
df = pd.DataFrame()
```

```python
# 불러오기를 시작할 날짜 설정
stockStartDate = '2018-01-01'

# 오늘 날짜를 기준으로 불러오기를 마칠 날짜 설정
today = datetime.today().strftime('%Y-%m-%d')

# 각 자산의 수정 종가를 불러오고 데이터 프레임을 저장
for t in Assets :
    df[t] = pdr.DataReader(t, stockStartDate , end=today)['Adj Close']
```

```python
# 잘 불러와졌는지 확인
df
```

```python
# 일별 수익률, 공분산 구하기
dailyReturn= df.pct_change()
cov_daily = dailyReturn.cov()

# 연도별 수익률, 공분산 구하기
annualReturn = dailyReturn.mean()*250
cov_annual = cov_daily * 250

# 수익률, 분산, 가중치를 담을 빈 리스트 생성
p_returns =[]
p_vol = []
p_weights = []

# 자산의 개수 구하기
n_assets = len(Assets)
 
# 생성할 포트폴리오 개수 설정
n_ports = 30000
```

```python
# 수익률, 분산 구하기
for s in range(n_ports) :
    wgt = np.random.random(n_assets)
    wgt /= np.sum(wgt)
    annual = np.dot(wgt, annualReturn)
    vol = np.sqrt(np.dot(wgt.T,np.dot(cov_annual,wgt)))
    p_returns.append(annual)
    p_vol.append(vol)
    p_weights.append(wgt)

p_returns = np.array(p_returns)
p_vol = np.array(p_vol)
```

```python
# Efficient Frontier 시각화
colors = p_returns / p_vol
plt.style.use('seaborn')
plt.scatter(p_vol, p_returns, c=colors, marker='o', cmap=mpl.cm.jet)
plt.xlabel('Vol(Std.Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()
```

```python
# 자산별 수익률 계산 
returns = df.pct_change()
returns
```

```python
# 자산별 공분산 계산
cov_matrix_annual = returns.cov() * 250
cov_matrix_annual
```

```python
# 포트폴리오 분산 계산
port_variance = np.dot(weights.T, np.dot(cov_matrix_annual, weights))
port_variance
```

```python
# 포트폴리오 표준편차(변동성) 계산
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
# 포트폴리오 연간 수익률 계산
portfolioSimpleAnnualReturn = np.sum(returns.mean()*weights) * 252
portfolioSimpleAnnualReturn
```

```python
# 포트폴리오 연간 기대수익률, 표준편차, 분산 정리
percent_var = str(round(port_variance, 2) * 100) + '%'
percent_vols = str(round(port_volatility, 2) * 100) + '%'
percent_ret = str(round(portfolioSimpleAnnualReturn, 2)*100)+'%'
print("Expected annual return : "+ percent_ret)
print('Annual volatility/standard deviation/risk : '+percent_vols)
print('Annual variance : '+percent_var)
```

### 포트폴리오 최적화

```python
# 포트폴리오 기대수익률 할당
mu = expected_returns.mean_historical_return(df)#returns.mean() * 250
# 포트폴리오 공분산 할당
S = risk_models.sample_cov(df) #Get the sample covariance matrix
```

```python
# EfficientFrontier 계산
ef = EfficientFrontier(mu, S)
# Sharpe ratio를 최대화하는 가중치 계산
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights() 
```

```python
# 포트폴리오 가중치 최적화 결과
print(cleaned_weights) 
```

```python
# 최적 포트폴리오의 return, risk 출력
ef.portfolio_performance(verbose=True)
```
