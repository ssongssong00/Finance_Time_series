# Data Load

```python
import pandas as pd
import numpy as np
data = pd.read_csv("QUARTERLY.csv")
```

# Data preprocessing

```python
data['IndProd_t-1'] = data['IndProd'].shift(1)
data['yt_inprod'] = np.log(data['IndProd']) - np.log(data['IndProd_t-1'])
```

```python
data
```

# a
coef const가 알파, coef ar.L1이 베타  
yt = 알파 + 베타 yt-1 + 엡실론 t

```python
import statsmodels.api as sm
model = sm.tsa.arima.ARIMA(data['yt_inprod'], order = (1,0,0))
```

```python
result = model.fit()
result.summary()
```
