# Data Load

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

data = pd.read_csv("QUARTERLY.csv")
```

# Data preprocessing

```python
data['yt_inprod'] = np.log(data['IndProd']) - np.log(data['IndProd'].shift(1))
```

```python
data.head(5)
```

# a
coef const가 알파, coef ar.L1이 베타  
yt = 알파 + 베타 yt-1 + 엡실론 t

```python
model = sm.tsa.arima.ARIMA(data['yt_inprod'], order = (1,0,0))
result = model.fit()
result.summary()
```
